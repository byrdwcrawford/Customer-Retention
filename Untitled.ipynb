```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encoding the categorical columns
ohe = OneHotEncoder(handle_unknown='error', sparse=False, drop='first')
columns_to_ohe = ['state', 'area code']
encoded = ohe.fit_transform(X_train[columns_to_ohe])

encoded
```

```python
# Get the unique categories for the columns you one-hot encoded
categories = ohe.categories_

# Construct column names based on the unique categories
column_names = []
for i, category_list in enumerate(categories):
    column_names.extend([f"{columns_to_ohe[i]}_{category}" for category in category_list])

# Trim column names to match the number of columns in the encoded array
column_names = column_names[:encoded.shape[1]]

# Convert the one-hot encoded array to a DataFrame with meaningful column names
encoded_df = pd.DataFrame(encoded, columns=column_names)

# Concatenate the one-hot encoded DataFrame with the original customer_df
X_train_preprocessed = pd.concat([customer_df.reset_index(drop=True), encoded_df], axis=1)

# Print the concatenated DataFrame
X_train_preprocessed.info()
X_train_preprocessed.head()



```
